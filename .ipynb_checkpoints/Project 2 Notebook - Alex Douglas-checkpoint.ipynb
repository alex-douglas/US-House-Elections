# Project 2: 2016 U.S. House of Representatives Elections

### Goal

The goal of this project is to see if we can predict the margin of victory in election results for the U.S. House of Representatives based on biographical information about the winning candidates.

### Reasoning

I would think that biographical information about individual politicans should have no impact on their margin of victory. Other variables are likely to drown out any influence marriage, children, or education might have on the outcome of the elections. However, the goal of this project was to develop our web scrapping skills in addition to learning about linear regression, so I am gathering biographical information for politicans since this data is available from their Wikipedia pages.

I will also get the margin of victory for each district in the 2014 elections, as a control variable to compare against the biographical variables. I would expect this to be a massive predictor of election results in 2016, since districts tend to vote similarly from year to year, and gerrymandering has made this feature of House elections more and more true. I will also get the date they assumed office and their party, which aren't biographical variables, and I would expect these two to potentially influence the margin of victory, though mabye not as much as the margin of victory in 2014.

### Project Process

1. Scrape the [wikipedia page](https://en.wikipedia.org/wiki/Current_members_of_the_United_States_House_of_Representatives) of the current list of Representatives, which will give their names, districts, and links to their specific Wikipedia pages.
2. Scrape the individual Wikipedia pages for each Representative, and get the following variables if available:
    - date assumed office
    - birthday
    - number of children
    - marital status
    - party
    - education (type of degree)
    - occupation
    - gender
3. Scrape the [2014 margins](https://ballotpedia.org/Margin_of_victory_analysis_for_the_2014_congressional_elections) from the ballotpedia.org website.
4. Scrape the [2016 margins](https://ballotpedia.org/United_States_House_of_Representatives_elections,_2016) from the ballotpedia.org website.
5. Combine all the collected data into one dataframe.
6. Clean the data.
6. Explore the data with linear regression techniques and determine the best model to predict the 2016 margins.

## Step 2: Scrape individual Representatives' Wikipedia pages for model variables

#### Let's first create our helper functions

```python
def generateRepUrl(rep_wiki_link: str):
    """Generates a url to a Representative's Wikipedia page, based on wiki link from df_links"""
    return 'https://en.wikipedia.org%s' % rep_wiki_link


def generateBeautifulSoup(rep_url):
    """Gets the soup object from the HTML"""
    response = requests.get(rep_url)
    
    if(response.status_code != 200):
        return -1
    else:
        rep_page = response.text
        return BeautifulSoup(rep_page,"lxml")
    

def generateInfobox(rep_soup):
    """Gets the infobox vcard object from the BeautifulSoup object, which contains most of the biographical information"""
    infobox = rep_soup.find(class_='infobox vcard')
    return infobox


def get_gender(rep_soup):
    """Get's the gender of the politican, based on the number of gendered
       pronouns in the leading and biographical sections of their Wikipedia
       page. There is probably a much better way to do this (scrape another website),
       but this is the best I could come up with given current time constraints"""
    male_pronouns = ['he', 'his', 'him']
    female_pronouns = ['she', 'hers', 'her', 'congresswoman']
    male_count = 0
    female_count = 0

    summary_str = str(rep_soup.find(class_='toc').find_all_previous('p')).lower()
    
    if rep_soup.find(id=lambda x: x and 'early' in x.lower()):
        early_life_str = str(rep_soup.find(id=lambda x: x and 'early' in x.lower()).find_next('p')).lower()
    elif rep_soup.find(id=lambda x: x and 'background' in x.lower()):
        early_life_str = str(rep_soup.find(id=lambda x: x and 'background' in x.lower()).find_next('p')).lower()
    else:
        early_life_str = ''

    for pronoun in male_pronouns:
        male_count += sum(1 for _ in re.finditer(r'\b%s\b' % re.escape(pronoun), summary_str))
        male_count += sum(1 for _ in re.finditer(r'\b%s\b' % re.escape(pronoun), early_life_str))

    for pronoun in female_pronouns:
        female_count += sum(1 for _ in re.finditer(r'\b%s\b' % re.escape(pronoun), summary_str))
        female_count += sum(1 for _ in re.finditer(r'\b%s\b' % re.escape(pronoun), early_life_str))

    if male_count > female_count:
        return 1
    elif female_count > male_count:
        return 0
    else:
        return None
    


def generateRepData(rep_soup, rep_name):
    """Parses the Representative's vcard for specific biographical information.
       This function is a result of many testing iterations"""
    
    rep_data = {}
    rep_data['name'] = rep_name
    degrees = []
    
    # some children counts use words instead of numbers, so create this mapping to convert words to numbers
    num_map = {'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10}
    
    infobox = generateInfobox(rep_soup)
    
    for table_row in infobox.find_all('tr'):
        for table_header in table_row.find_all('th'):
            
            # Political Party
            if 'Political party' in table_header.text:
                rep_data['party'] = table_header.next_element.next_element.next_element.text
            
            #Marriage Status
            elif 'Spouse' in table_header.text:
                spouse = table_header.next_element.next_element.next_element.text.lower()
                if 'divorced' in spouse or spouse == 'none':
                    rep_data['married'] = 0
                else: rep_data['married'] = 1
                    
            #Number of Children
            elif 'Children' in table_header.text:
                children = table_header.next_element.next_element.next_element.text.lower()
                if children.isdigit():
                    rep_data['children'] = int(children)
                elif '\n' in children:
                    rep_data['children'] = len(str.splitlines(children))
                else:
                    children = children.split(' ')
                    if children[0] in num_map:
                        rep_data['children'] = num_map[children[0]]
                    elif any(char.isdigit() for char in children[0]):
                        rep_data['children'] = int(children[0][:1])
                    else:
                        try:
                            rep_data['children'] = int(children[0])
                        except ValueError:
                            continue
                            
            #Education (type of degree)
            elif table_header.text == 'Education':
                for item in table_header.next_element.next_element.next_element.find_all('a'):
                    if len(item.text) < 7:
                        degrees.append(item.text.replace('.', ''))
                        
            #Occupation
            elif 'Occupation' in table_header.text:
                rep_data['occupation'] = table_header.next_element.next_element.next_element.text
                
            #Date assumed office
            elif 'U.S. House of Representatives' in table_header.text:
                if table_header.find_next('tr').find_next('tr').find('br'):
                    rep_data['assumed_office'] = table_header.find_next('tr').find_next('tr').find('br').next_sibling.strip()
        for table_data in table_row.find_all('td'):
            
            #Birthday
            if table_data.find(class_='bday'):
                rep_data['birthday'] = table_data.find(class_='bday').text
                
    # get gender
    rep_data['male'] = get_gender(rep_soup)
    
    # Assign NoneType to variable if it wasn't found on the Wikipedia page
    if 'assumed_office' not in rep_data: 
        rep_data['assumed_office'] = None
    if 'birthday' not in rep_data: 
        rep_data['birthday'] = None
    if 'children' not in rep_data: 
        rep_data['children'] = None
    if 'married' not in rep_data: 
        rep_data['married'] = None
    if 'party' not in rep_data: 
        rep_data['party'] = None
    if 'occupation' not in rep_data: 
        rep_data['occupation'] = None
        
    if degrees:
        rep_data['education'] = degrees
    else:
        rep_data['education'] = None
    
    return rep_data
```

#### Now let's actually create the dataframe

```python
# generate main dataframe
column_values = ['name', 'assumed_office', 'birthday', 'children', 'married', 'party', 'education', 'occupation', 'male']
df_rep_data = pd.DataFrame(columns=column_values, index=range(435))
```

## Step 3: Scrape for 2014 margin data

```python
# get the html from the webpage
margin_2014_url = 'https://ballotpedia.org/Margin_of_victory_analysis_for_the_2014_congressional_elections'
margin_2014_response = requests.get(margin_2014_url)
# confirm request success
print(margin_2014_response.status_code)
# convert the response to a beautifulSoup object
if margin_2014_response.status_code == 200:
    margin_2014_page = margin_2014_response.text
    margin_2014_soup = BeautifulSoup(margin_2014_page,"lxml")
```

```python
# let's check out the soup
print(margin_2014_soup.prettify())
```

```python
# get the table containing the margins of victory
margin_table_2014 = margin_2014_soup.find(text=re.compile('^District$')).parent.parent.parent.parent
```

```python
margin_2014_list = [['district', 'margin_2014']] # list to eventually convert to dataframe

for table_data in margin_table_2014.find_all('td'):
    if 'District' in table_data.text:
        margin_2014_list.append([table_data.text.replace('District ', '').replace('District', '')
                                                                         .replace("'s At-Large", ' At Large')
                                                                         .replace(' Runoff Election', '')
                                                                         .replace(',', '').strip(), 
                                 float(table_data.next_sibling.next_sibling.next_sibling.next_sibling.text.strip('%'))])
```

```python
# replace 100% values (which means election was uncontested) with None
for i, row in enumerate(margin_2014_list):
    if margin_2014_list[i][1] == 100.0:
        margin_2014_list[i][1] = None
```

```python
# get rid of Special Elections, they are unnessary
for i, row in enumerate(margin_2014_list):
    if 'Special Election' in margin_2014_list[i][0]:
        del margin_2014_list[i]
```

```python
df_2014_margins = pd.DataFrame(margin_2014_list[1:], columns=margin_2014_list[:1][0])
```

```python
# create helper functions to convert birthday and assumed_office to years, as of 2016 elections
def convert_birthday(row):
    timedelta = datetime.date(2016, 11, 8) - row.birthday.date()
    return timedelta.days / 365.2425

def convert_assumed_office(row):
    timedelta = datetime.date(2017, 11, 8) - row.assumed_office.date()
    return timedelta.days / 365.2425
```

##### Let's run a basic linear regression to see how coefficients and significance look

```python
# create our y and X matrices with the help of patsy
y, X = patsy.dmatrices('margin_2016 ~ years_office + age + children + married + republican + has_jd + male + children_found + marriage_found + education_found + occupation_found + all_data_found + margin_2014', data=df_lr, return_type="dataframe")
model = sm.OLS(y, X)
fit = model.fit()
fit.summary()
```

```python
# let's eliminate the most insignificant features one by one and see if other features become significant as elmination occurs
def testElimFeatures(dataframe: pd.DataFrame, target: str, features: list, significance: float):
    sig_feature_count = 0
    dropped_feature = ''
    
    while len(features) >= sig_feature_count:
        model_str = target + ' ~ ' + ' + '.join(features)
        y_elim, X_elim = patsy.dmatrices(model_str, data=dataframe, return_type="dataframe")
        model_elim = sm.OLS(y_elim, X_elim)
        fit_elim = model_elim.fit()
        sig_features = list(fit_elim.pvalues[fit_elim.pvalues <= significance].index)
        sig_feature_count = len(sig_features)
        print(dropped_feature, ':', ' ' * (18 - len(dropped_feature)), sig_features, sep='')
        dropped_feature = fit_elim.pvalues[fit_elim.pvalues == max(fit_elim.pvalues)].index[0]
        features.remove(dropped_feature)
```

```python
testElimFeatures(df_lr, 'margin_2016', list(df_lr.columns.drop('margin_2016')), 0.10)
```

None of the features are statistically significant at the 0.10 level, except for the republican dummy and margin_2014. So we'll use a very small LR model with just two features (+ an intercept), and see if we can improve our R<sup>2</sup>.
##### Can we improve the R<sup>2</sup>

```python
# let's get a base model to compare against
cross_valiator = KFold(n_splits=5, shuffle=True, random_state=42)
y_base, X_base = patsy.dmatrices('margin_2016 ~ republican + margin_2014', data=df_lr, return_type="dataframe")

np.median(cross_val_score(LinearRegression(), X_base, y_base, cv=cross_valiator))
```

Does regularization help?

```python
ridge_cv_model = RidgeCV(cv=cross_valiator, alphas=np.arange(0.1, 50.0, 0.1), normalize=True)
ridge_cv_model.fit(X_base, y_base)

ridge_model = Ridge(alpha=ridge_cv_model.alpha_)
print(np.median(cross_val_score(ridge_model, X_base, y_base, cv=cross_valiator)), ridge_cv_model.alpha_)
```

```python
lasso_cv_model = LassoCV(cv=cross_valiator, alphas=np.arange(0.001, 0.020, 0.001), normalize=True)
lasso_cv_model.fit(X_base, y_base.values.ravel())

lasso_model = Lasso(alpha=lasso_cv_model.alpha_)
print(np.median(cross_val_score(lasso_model, X_base, y_base, cv=cross_valiator)), lasso_cv_model.alpha_)
```

Regularization doesn't seem to help, perhaps raising the model to a higher order polynomial will improve the predictive power:

```python
# get the train/test split
X_poly_train, X_poly_test, y_poly_train, y_poly_test = train_test_split(df_lr[['margin_2014', 'republican']], df_lr['margin_2016'], test_size=0.2)
```

```python
poly_model = make_pipeline(PolynomialFeatures(2), LinearRegression())
poly_model.fit(X_poly_train, y_poly_train)
r2_score(y_poly_test, poly_model.predict(X_poly_test))
```

Raising the model to the second power seems to quite dramatically improve the predictive power of the model. Let's see if we can get the p-values and coefficients too from statsmodels.

```python
df_lr_poly = df_lr[['margin_2016', 'margin_2014', 'republican']]
df_lr_poly['margin_2014_squared'] = df_lr_poly['margin_2014']**2
df_lr_poly['republican_squared'] = df_lr_poly['republican']**2
df_lr_poly['margin_2014_republican'] = df_lr_poly['margin_2014'] * df_lr_poly['republican']
```

```python
df_lr_poly.sample(4)
```

```python
y_poly, X_poly = patsy.dmatrices('margin_2016 ~ republican + margin_2014 + margin_2014_squared + margin_2014_republican', data=df_lr_poly, return_type="dataframe")
model_poly = sm.OLS(y_poly, X_poly)
fit_poly = model_poly.fit()
fit_poly.summary()
```

```python
# everything is significant except for the interaction between margin_2014 and republican, so let's drop that term
y_poly_2, X_poly_2 = patsy.dmatrices('margin_2016 ~ republican + margin_2014 + margin_2014_squared', data=df_lr_poly, return_type="dataframe")
model_poly_2 = sm.OLS(y_poly_2, X_poly_2)
fit_poly_2 = model_poly_2.fit()
fit_poly_2.summary()
```

Let's rerun the regularization models with this new polynomial term built in:

```python
ridge_cv_model_poly = RidgeCV(cv=cross_valiator, alphas=np.arange(0.1, 50.0, 0.1), normalize=True)
ridge_cv_model_poly.fit(X_poly_2, y_poly_2)

ridge_model_poly = Ridge(alpha=ridge_cv_model_poly.alpha_)
print(np.median(cross_val_score(ridge_model_poly, X_poly_2, y_poly_2, cv=cross_valiator)), ridge_cv_model_poly.alpha_)
```

```python
lasso_cv_model_poly = LassoCV(cv=cross_valiator, alphas=np.arange(0.0001, 0.0020, 0.0001), normalize=True)
lasso_cv_model_poly.fit(X_poly_2, y_poly_2.values.ravel())

lasso_model_poly = Lasso(alpha=lasso_cv_model_poly.alpha_)
print(np.median(cross_val_score(lasso_model_poly, X_poly_2, y_poly_2, cv=cross_valiator)), lasso_cv_model_poly.alpha_)
```

There appears to be no real difference between the two types of regularization, so there likely isn't any value in also computing an elastic net  regularization. Let's instead compute a non-regularized cross-validated score on the polynomial data and see if that improves the original R<sup>2</sup> of 0.378.

```python
np.median(cross_val_score(LinearRegression(), X_poly_2, y_poly_2, cv=cross_valiator))
```

## Conclusion

None of the biographical variables were statistically signficant; the only two statistically significant predictors of 2016 margins of victory were the party affiliation of the candidates and the district's margin of victory in 2014.

Adding a polynomial feature of "2014 margin squared" improved the predictive power of the model by a couple of points, although regularizaiton had no impact. Therefore the final model consists of three features: republican (dummy), 2014 margin, 2014 margin squared. Taking the coefficents given from the statsmodels summary, we have a final equation of:

$y = 13.3982 - 8.7686(republican) +0.9414(\textit{2014_margin}) - 0.0038(\textit{2014_margin})^2$
